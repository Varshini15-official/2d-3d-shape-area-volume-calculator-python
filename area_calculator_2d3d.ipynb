```python
from tkinter import *
import math

root=Tk()
root.geometry("600x450")
root.configure(background="Pink")
root.title("Area Calculator")

bg=PhotoImage(file="blk bgn.png")
bglabel=Label(root,image=bg)
bglabel.place(x=0,y=0,relwidth=1,relheight=1)

def twod():
    root.destroy()
    groot=Tk()
    groot.geometry("400x450")
    groot.configure(background="Pink")
    groot.title("2-D Figures")
    
    def circle():
        groot.destroy()
        def calc(r):
            return math.pi*r*r
        def clickarea():
            area.set(calc(r.get()))
            draw_circle(float(r.get()))
        broot=Tk()
        broot.geometry("400x400")
        broot.configure(background="Light Green")
        broot.title("2-D Figures")
        r=IntVar()
        Label(broot,text="Enter The Radius Of the Circle").pack()
        Entry(broot,textvariable=r).pack()
        area=IntVar()
        Label(broot,text="The Area Of The Circle Is:").pack()
        Label(broot,textvariable=area).pack()
        Button(broot,text="Calculate Area",command=clickarea).pack()
        canvas = Canvas(broot, width=300, height=300, bg="white")
        canvas.pack()
        def draw_circle(r):
            canvas.delete("all")
            x = 300 / 2
            y = 300 / 2
            canvas.create_oval(x - r, y - r, x + r, y + r, outline="black", fill="light blue")
        broot.mainloop()
    
    
    def square():
        groot.destroy()
        def calc(s):
            return s*s
        def clickarea():
            try:
                area.set(calc(s.get()))
                draw_square(float(s.get()))
            except ValueError:
                messagebox.showerror("Error", "Please enter valid numbers.")
        broot=Tk()
        broot.geometry("400x400")
        broot.configure(background="Light Green")
        broot.title("Square")
        s=IntVar()
        Label(broot,text="Enter The Side Of The Square").pack()
        Entry(broot,textvariable=s).pack()
        area=IntVar()
        Label(broot,text="The Area Of The Square Is:").pack()
        Label(broot,textvariable=area).pack()
        Button(broot,text="Calculate Area",command=clickarea).pack()
        canvas = Canvas(broot, width=300, height=300, bg="white")
        canvas.pack()
        def draw_square(s):
            canvas.delete("all")
            x1 = 150 - s / 2
            y1 = 150 - s / 2
            x2 = 150 + s / 2
            y2 = 150 - s / 2
            x3 = 150 + s / 2
            y3 = 150 + s / 2
            x4 = 150 - s / 2
            y4 = 150 + s / 2
            canvas.create_polygon(x1, y1, x2, y2, x3, y3, x4, y4, outline="black", fill="light blue")
        broot.mainloop()
    
    
    def rect():
        groot.destroy()
        def calc(l, b):
            return l * b
        def clickarea():
            try:
                area.set(calc(float(l.get()), float(b.get())))
                draw_rectangle(float(l.get()), float(b.get()))
            except ValueError:
                messagebox.showerror("Error", "Please enter valid numbers.")
        broot=Tk()
        broot.geometry("400x400")
        broot.configure(background="Light Green")
        broot.title("Rectangle")
        l = IntVar()
        b = IntVar()
        area = IntVar()
        Label(broot, text="Enter The Length Of The Rectangle").pack()
        Entry(broot, textvariable=l).pack()
        Label(broot, text="Enter The Breadth Of The Rectangle").pack()
        Entry(broot, textvariable=b).pack()
        area=IntVar()
        Label(broot, text="The Area Of The Rectangle Is:").pack()
        Label(broot, textvariable=area).pack()
        Button(broot, text="Calculate Area", command=clickarea).pack()
        canvas = Canvas(broot, width=300, height=300, bg="white")
        canvas.pack()
        
        def draw_rectangle(l, b):
            canvas.delete("all")
            x1 = (150 - l) / 2
            y1 = (150 - b) / 2
            x2 = (150 + l) / 2
            y2 = (150 + b) / 2
            canvas.create_polygon(x1, y1, x2, y1, x2, y2, x1, y2, outline="black", fill="light blue")
        broot.mainloop()

    def rhombus():
        groot.destroy()
        def calc(d,dd):
            return 0.5*d*dd
        def clickarea():
            try:
                area.set(calc(d.get(),dd.get()))
                draw_rhombus(float(d.get()), float(dd.get()))
            except ValueError:
                messagebox.showerror("Error", "Please enter valid numbers.")
        broot=Tk()
        broot.geometry("400x400")
        broot.configure(background="Light Green")
        broot.title("Rhombus")
        d=IntVar()
        dd=IntVar()
        Label(broot,text="Enter The First Diagonal Measure").pack()
        Entry(broot,textvariable=d).pack()
        Label(broot,text="Enter The Second Diagonal Measure").pack()
        Entry(broot,textvariable=dd).pack()
        area=IntVar()
        Label(broot,text="The Area Of The Rhombus Is:").pack()
        Label(broot,textvariable=area).pack()
        Button(broot,text="Calculate Area",command=clickarea).pack()
        canvas = tk.Canvas(broot, width=300, height=300, bg="white")
        canvas.pack()
           
        def draw_rhombus(d, dd):
            canvas.delete("all")
            x1 = 300 / 2 - d / 2
            y1 = 300 / 2
            x2 = 300 / 2 + d / 2
            y2 = 300 / 2
            x3 = 300 / 2 + dd / 2
            y3 = 300 / 2 - d / 2
            x4 = 300 / 2 - dd / 2
            y4 = 300 / 2 - d / 2
            canvas.create_polygon(x1, y1, x2, y2, x3, y3, x4, y4, outline="black", fill="light green")
        broot.mainloop()

    def triangle():
        groot.destroy()
        def calc(b,h):
            return 0.5*b*h
        def clickarea():
            try:
                area.set(calc(b.get(),h.get()))
                draw_triangle(float(b.get()), float(h.get()))
            except ValueError:
                messagebox.showerror("Error", "Please enter valid numbers.")
        broot=Tk()
        broot.geometry("400x400")
        broot.configure(background="Light Green")
        broot.title("Triangle")
        b=IntVar()
        h=IntVar()
        Label(broot,text="Enter The Base Of The Triangle").pack()
        Entry(broot,textvariable=b).pack()
        Label(broot,text="Enter The Height Of The Triangle").pack()
        Entry(broot,textvariable=h).pack()
        area=IntVar()
        Label(broot,text="The Area Of The Triangle Is:").pack()
        Label(broot,textvariable=area).pack()
        Button(broot,text="Calculate Area",command=clickarea).pack()
        canvas = tk.Canvas(broot, width=300, height=300, bg="white")
        canvas.pack()
        
        def draw_triangle(b, h):
            canvas.delete("all")
            x1 = 300 / 2 - b / 2
            y1 = 300 / 2 - h / 2
            x2 = 300 / 2 + b / 2
            y2 = 300 / 2 - h / 2
            x3 = 300 / 2
            y3 = 300 / 2 + h / 2
            canvas.create_polygon(x1, y1, x2, y2, x3, y3, outline="black", fill="light green")
        broot.mainloop()    
        
    def parallel():
        groot.destroy()
        def calc(b, h):
            return b * h
        def clickarea():
            try:
                area.set(calc(float(b.get()), float(h.get())))
                draw_parallelogram(float(b.get()), float(h.get()))
            except ValueError:
                messagebox.showerror("Error", "Please enter valid numbers.")
        broot=Tk()
        broot.geometry("400x400")
        broot.configure(background="Light Green")
        broot.title("Parallelogram")
        b=IntVar()
        h=IntVar()
        Label(broot,text="Enter The Base Of The Parallelogram").pack()
        Entry(broot,textvariable=b).pack()
        Label(broot,text="Enter The Height Of The Parallelogram").pack()
        Entry(broot,textvariable=h).pack()
        area=IntVar()
        Label(broot,text="The Area Of The Parallelogram Is:").pack()
        Label(broot,textvariable=area).pack()
        Button(broot,text="Calculate Area",command=clickarea).pack()
        canvas = tk.Canvas(broot, width=300, height=300, bg="white")
        canvas.pack()
        
        def draw_parallelogram(b, h):
            canvas.delete("all")
            x1 = 150 - h / 2
            y1 = 150 - b / 2
            x2 = 150 + h / 2
            y2 = 150 + b / 2
            x3 = x1
            y3 = y2
            x4 = x2
            y4 = y1
            canvas.create_polygon(x1, y1, x2, y2, x3, y3, x4, y4, outline="black", fill="light blue")
        broot.mainloop()

    def trap():
        groot.destroy()
        def calc(s,ss,h):
            return 0.5*s*ss*h
        def clickarea():
            area.set(calc(s.get(),ss.get(),h.get()))
        broot=Tk()
        broot.geometry("400x400")
        broot.configure(background="Light Green")
        broot.title("Trapezium")
        s=IntVar()
        ss=IntVar()
        h=IntVar()
        Label(broot,text="Enter The First Parallel Side").pack()
        Entry(broot,textvariable=s).pack()
        Label(broot,text="Enter The Second Parallel Side").pack()
        Entry(broot,textvariable=ss).pack()
        Label(broot,text="Enter The Height Of Trapezium").pack()
        Entry(broot,textvariable=h).pack()
        area=IntVar()
        Label(broot,text="The Area Of The Trapezium Is:").pack()
        Label(broot,textvariable=area).pack()
        Button(broot,text="Calculate Area",command=clickarea).pack()
        canvas = Canvas(broot, width=300, height=300, bg="white")
        canvas.pack()
        def draw_trapezium(s, ss, h):
            canvas.delete("all")
            x1, y1 = 150 - s / 2, 150 - h / 2
            x2, y2 = 150 + s / 2, 150 - h / 2
            x3, y3 = 150 + ss / 2, 150 + h / 2
            x4, y4 = 150 - ss / 2, 150 + h / 2
            canvas.create_polygon(x1, y1, x2, y2, x3, y3, x4, y4, outline="black", fill="light green")
        broot.mainloop()

    def sect():
        groot.destroy()
        def calc(theta,r):
            return (theta/360)*math.pi*r*r
        def clickarea():
            area.set(calc(theta.get(),r.get()))
        broot=Tk()
        broot.geometry("400x400")
        broot.configure(background="Light Green")
        broot.title("Sector Of Circle")
        theta=IntVar()
        r=IntVar()
        Label(broot,text="Enter The Angle In Degrees").pack()
        Entry(broot,textvariable=theta).pack()
        Label(broot,text="Enter The Radius Of The Sector").pack()
        Entry(broot,textvariable=r).pack()
        area=IntVar()
        Label(broot,text="The Area Of The Sector Is:").pack()
        Label(broot,textvariable=area).pack()
        Button(broot,text="Calculate Area",command=clickarea).pack()
        canvas = Canvas(broot, width=300, height=300, bg="white")
        canvas.pack()
        def draw_sector(theta, r):
            canvas.delete("all")
            canvas.delete("all")
            x1, y1 = 150, 150
            x2, y2 = x1 + r * math.cos(math.radians(0)), y1 + r * math.sin(math.radians(0))
            x3, y3 = x1 + r * math.cos(math.radians(theta)), y1 + r * math.sin(math.radians(theta))
            canvas.create_arc(50, 50, 250, 250, start=0, extent=theta, outline="black", fill="light green")
            canvas.create_line(x1, y1, x2, y2, fill="black")
        broot.mainloop()

    cbutton=Button(groot,text="Circle",bg="#E070F6",command=circle,width=10,height=2)
    cbutton.place(x=50,y=50)
    sbutton=Button(groot,text="Square",bg="#E070F6",command=square,width=10,height=2).place(x=300,y=50)
    rbutton=Button(groot,text="Rectangle",bg="#E070F6",command=rect,width=10,height=2).place(x=50,y=150)
    rhbutton=Button(groot,text="Rhombus",bg="#E070F6",command=rhombus,width=10,height=2).place(x=300,y=150)
    tbutton=Button(groot,text="Triangle",bg="#E070F6",command=triangle,width=10,height=2).place(x=50,y=250)
    pbutton=Button(groot,text="Parallelogram",bg="#E070F6",command=parallel,width=10,height=2).place(x=300,y=250)
    trbutton=Button(groot,text="Trapezium",bg="#E070F6",command=trap,width=10,height=2).place(x=50,y=350)
    sebutton=Button(groot,text="Sector Of Circle",bg="#E070F6",command=sect,width=10,height=2).place(x=300,y=350)

    groot.mainloop()

def thrd():
    root.destroy()
    groot=Tk()
    groot.geometry("400x450")
    groot.configure(background="light slate gray")
    groot.title("3-D Figures")    
    def cube():
        groot.destroy()
        def calc(s):
            return s ** 3
        def apply_rotation_matrix(point, matrix):
            x, y, z = point
            new_x = matrix[0][0] * x + matrix[0][1] * y + matrix[0][2] * z
            new_y = matrix[1][0] * x + matrix[1][1] * y + matrix[1][2] * z
            new_z = matrix[2][0] * x + matrix[2][1] * y + matrix[2][2] * z
            return new_x, new_y, new_z

        def draw_cube(s, x_angle, y_angle, z_angle):
            canvas.delete("all")
            vertices = [
                [-s / 2, -s / 2, -s / 2],
                [s / 2, -s / 2, -s / 2],
                [s / 2, s / 2, -s / 2],
                [-s / 2, s / 2, -s / 2],
                [-s / 2, -s / 2, s / 2],
                [s / 2, -s / 2, s / 2],
                [s / 2, s / 2, s / 2],
                [-s / 2, s / 2, s / 2]
            ]
            edges = [
                [0, 1], [1, 2], [2, 3], [3, 0],
                [4, 5], [5, 6], [6, 7], [7, 4],
                [0, 4], [1, 5], [2, 6], [3, 7]
            ]
            rotation_matrix_x = [
                [1, 0, 0],
                [0, math.cos(math.radians(x_angle)), -math.sin(math.radians(x_angle))],
                [0, math.sin(math.radians(x_angle)), math.cos(math.radians(x_angle))]
            ]
            rotation_matrix_y = [
                [math.cos(math.radians(y_angle)), 0, math.sin(math.radians(y_angle))],
                [0, 1, 0],
                [-math.sin(math.radians(y_angle)), 0, math.cos(math.radians(y_angle))]
            ]
            rotation_matrix_z = [
                [math.cos(math.radians(z_angle)), -math.sin(math.radians(z_angle)), 0],
                [math.sin(math.radians(z_angle)), math.cos(math.radians(z_angle)), 0],
                [0, 0, 1]
            ]
            rotated_vertices = []
            for vertex in vertices:
                rotated_vertex = apply_rotation_matrix(vertex, rotation_matrix_x)
                rotated_vertex = apply_rotation_matrix(rotated_vertex, rotation_matrix_y)
                rotated_vertex = apply_rotation_matrix(rotated_vertex, rotation_matrix_z)
                rotated_vertices.append(rotated_vertex)
            projected_points = []
            for vertex in rotated_vertices:
                distance = 5  
                x = vertex[0] * distance + canvas.winfo_reqwidth() / 2
                y = vertex[1] * distance + canvas.winfo_reqheight() / 2
                projected_points.append([x, y])
            for edge in edges:
                canvas.create_line(projected_points[edge[0]][0], projected_points[edge[0]][1],
                                   projected_points[edge[1]][0], projected_points[edge[1]][1], fill="black")
        def clickvolume():
            try:
                volume.set(calc(s.get()))
                x_angle_val = int(x_angle.get())
                y_angle_val = int(y_angle.get())
                z_angle_val = int(z_angle.get())
                draw_cube(s.get(), x_angle_val, y_angle_val, z_angle_val)
            except ValueError:
                messagebox.showerror("Error", "Please enter valid numbers.")
        broot = Tk()
        broot.geometry("1000x1000")
        broot.configure(background="Light Blue")
        broot.title("3D Cube")
        s = IntVar()
        x_angle = IntVar()
        y_angle = IntVar()
        z_angle = IntVar()
        Label(broot, text="Enter The Side Of the Cube").pack()
        Entry(broot, textvariable=s).pack()
        Label(broot, text="Enter The Rotation Angle For X-Axis").pack()
        Entry(broot, textvariable=x_angle).pack()
        Label(broot, text="Enter The Rotation Angle For Y-Axis").pack()
        Entry(broot, textvariable=y_angle).pack()
        Label(broot, text="Enter The Rotation Angle For Z-Axis").pack()
        Entry(broot, textvariable=z_angle).pack()
        volume = IntVar()
        Label(broot, text="The Volume Of the Cube Is:").pack()
        Label(broot, textvariable=volume).pack()
        Button(broot, text="Calculate Volume", command=clickvolume).pack()
        canvas = Canvas(broot, width=750, height=750, bg="white")
        canvas.pack()
        broot.mainloop()

    def cuboid():
        groot.destroy()
        def calc(l, b, h):
            return l * b * h
        def apply_rotation_matrix(point, matrix):
            x, y, z = point
            new_x = matrix[0][0] * x + matrix[0][1] * y + matrix[0][2] * z
            new_y = matrix[1][0] * x + matrix[1][1] * y + matrix[1][2] * z
            new_z = matrix[2][0] * x + matrix[2][1] * y + matrix[2][2] * z
            return new_x, new_y, new_z
        def draw_cuboid(l, b, h, x_angle, y_angle, z_angle):
            canvas.delete("all")
            vertices = [
                [0, 0, 0], [l, 0, 0], [l, h, 0], [0, h, 0],
                [0, 0, b], [l, 0, b], [l, h, b], [0, h, b]
            ]
            for i in range(len(vertices)):
                vertices[i][0] -= l / 2
                vertices[i][1] -= h / 2
                vertices[i][2] -= b / 2
            edges = [
                [0, 1], [1, 2], [2, 3], [3, 0],
                [4, 5], [5, 6], [6, 7], [7, 4],
                [0, 4], [1, 5], [2, 6], [3, 7]
            ]
            rx = math.radians(x_angle)
            ry = math.radians(y_angle)
            rz = math.radians(z_angle)
            rotation_matrix_x = [
                [1, 0, 0],
                [0, math.cos(rx), -math.sin(rx)],
                [0, math.sin(rx), math.cos(rx)]
            ]
            rotation_matrix_y = [
                [math.cos(ry), 0, math.sin(ry)],
                [0, 1, 0],
                [-math.sin(ry), 0, math.cos(ry)]
            ]
            rotation_matrix_z = [
                [math.cos(rz), -math.sin(rz), 0],
                [math.sin(rz), math.cos(rz), 0],
                [0, 0, 1]
            ]
            rotated_vertices = []
            for vertex in vertices:
                v = apply_rotation_matrix(vertex, rotation_matrix_x)
                v = apply_rotation_matrix(v, rotation_matrix_y)
                v = apply_rotation_matrix(v, rotation_matrix_z)
                rotated_vertices.append(v)
            projected = []
            for x, y, z in rotated_vertices:
                distance = 5
                px = x * distance + canvas.winfo_reqwidth() / 2
                py = y * distance + canvas.winfo_reqheight() / 2
                projected.append([px, py])
            for edge in edges:
                canvas.create_line(projected[edge[0]][0], projected[edge[0]][1],
                                   projected[edge[1]][0], projected[edge[1]][1], fill="black")
        def clickvolume():
            try:
                l_val, b_val, h_val = l.get(), b.get(), h.get()
                x_angle_val = x_angle.get()
                y_angle_val = y_angle.get()
                z_angle_val = z_angle.get()
                volume.set(calc(l_val, b_val, h_val))
                draw_cuboid(l_val, b_val, h_val, x_angle_val, y_angle_val, z_angle_val)
            except:
                volume.set("Invalid input")
        root = Tk()
        root.geometry("1000x1000")
        root.configure(bg="lightblue")
        root.title("Cuboid")
        l, b, h = IntVar(), IntVar(), IntVar()
        x_angle, y_angle, z_angle = IntVar(), IntVar(), IntVar()
        volume = IntVar()
        Label(root, text="Length").pack()
        Entry(root, textvariable=l).pack()
        Label(root, text="Breadth").pack()
        Entry(root, textvariable=b).pack()
        Label(root, text="Height").pack()
        Entry(root, textvariable=h).pack()
        Label(root, text="X Rotation Angle").pack()
        Entry(root, textvariable=x_angle).pack()
        Label(root, text="Y Rotation Angle").pack()
        Entry(root, textvariable=y_angle).pack()
        Label(root, text="Z Rotation Angle").pack()
        Entry(root, textvariable=z_angle).pack()
        Button(root, text="Calculate Volume & Draw", command=clickvolume).pack()
        Label(root, text="Volume:").pack()
        Label(root, textvariable=volume).pack()
        canvas = Canvas(root, width=750, height=750, bg="white")
        canvas.pack()
        broot.mainloop()

    def sphere():
        groot.destroy()
        def apply_rotation_matrix(point, matrix):
            x, y, z = point
            new_x = matrix[0][0]*x + matrix[0][1]*y + matrix[0][2]*z
            new_y = matrix[1][0]*x + matrix[1][1]*y + matrix[1][2]*z
            new_z = matrix[2][0]*x + matrix[2][1]*y + matrix[2][2]*z
            return new_x, new_y, new_z
        def draw_sphere(r, x_angle, y_angle, z_angle):
            canvas.delete("all")
            points = []
            # Create points on sphere using latitude and longitude
            lat_steps = 12  # number of latitude lines
            lon_steps = 24  # number of longitude lines
            for i in range(lat_steps + 1):
                lat = math.pi * (i / lat_steps - 0.5)  # from -pi/2 to +pi/2
                for j in range(lon_steps + 1):
                    lon = 2 * math.pi * (j / lon_steps)  # from 0 to 2pi
                    x = r * math.cos(lat) * math.cos(lon)
                    y = r * math.sin(lat)
                    z = r * math.cos(lat) * math.sin(lon)
                    # Rotate point
                    v = (x, y, z)
                    v = apply_rotation_matrix(v, rotation_matrix_x)
                    v = apply_rotation_matrix(v, rotation_matrix_y)
                    v = apply_rotation_matrix(v, rotation_matrix_z)
                    points.append(v)
            # Project and draw latitude lines
            for i in range(lat_steps + 1):
                line_points = points[i*(lon_steps+1):(i+1)*(lon_steps+1)]
                proj_points = []
                for p in line_points:
                    px = p[0]*scale + center_x
                    py = -p[1]*scale + center_y
                    proj_points.append((px, py))
                for k in range(len(proj_points) - 1):
                    canvas.create_line(proj_points[k][0], proj_points[k][1],
                                       proj_points[k+1][0], proj_points[k+1][1], fill="black")
            # Project and draw longitude lines
            for j in range(lon_steps + 1):
                line_points = []
                for i in range(lat_steps + 1):
                    idx = i*(lon_steps+1) + j
                    p = points[idx]
                    px = p[0]*scale + center_x
                    py = -p[1]*scale + center_y
                    line_points.append((px, py))
                for k in range(len(line_points) -1):
                    canvas.create_line(line_points[k][0], line_points[k][1],
                                       line_points[k+1][0], line_points[k+1][1], fill="black")
        def calc_volume(r):
            return (4/3) * math.pi * (r**3)
        def clickvolume():
            try:
                radius_val = float(radius.get())
                volume.set(round(calc_volume(radius_val), 3))
                global rotation_matrix_x, rotation_matrix_y, rotation_matrix_z
                x_val = float(x_angle.get())
                y_val = float(y_angle.get())
                z_val = float(z_angle.get())
                rotation_matrix_x = [
                    [1, 0, 0],
                    [0, math.cos(math.radians(x_val)), -math.sin(math.radians(x_val))],
                    [0, math.sin(math.radians(x_val)), math.cos(math.radians(x_val))]
                ]
                rotation_matrix_y = [
                    [math.cos(math.radians(y_val)), 0, math.sin(math.radians(y_val))],
                    [0, 1, 0],
                    [-math.sin(math.radians(y_val)), 0, math.cos(math.radians(y_val))]
                ]
                rotation_matrix_z = [
                    [math.cos(math.radians(z_val)), -math.sin(math.radians(z_val)), 0],
                    [math.sin(math.radians(z_val)), math.cos(math.radians(z_val)), 0],
                    [0, 0, 1]
                ]
                draw_sphere(radius_val, x_val, y_val, z_val)
            except ValueError:
                messagebox.showerror("Error", "Please enter valid numbers.")
        root = Tk()
        root.geometry("1000x1000")
        root.title("Rotating Sphere Wireframe")
        root.configure(bg="Light Blue")
        radius = StringVar()
        x_angle = StringVar()
        y_angle = StringVar()
        z_angle = StringVar()
        volume = StringVar()
        Label(root, text="Enter Radius of Sphere").pack()
        Entry(root, textvariable=radius).pack()
        Label(root, text="Rotation Angle X (degrees)").pack()
        Entry(root, textvariable=x_angle).pack()
        Label(root, text="Rotation Angle Y (degrees)").pack()
        Entry(root, textvariable=y_angle).pack()
        Label(root, text="Rotation Angle Z (degrees)").pack()
        Entry(root, textvariable=z_angle).pack()
        Label(root, text="Volume of Sphere:").pack()
        Label(root, textvariable=volume).pack()
        Button(root, text="Calculate & Draw Sphere", command=clickvolume).pack()
        canvas = Canvas(root, width=750, height=750, bg="white")
        canvas.pack(pady=10)
        center_x, center_y = 250, 250
        scale = 5 
        rotation_matrix_x = [[1,0,0],[0,1,0],[0,0,1]]
        rotation_matrix_y = [[1,0,0],[0,1,0],[0,0,1]]
        rotation_matrix_z = [[1,0,0],[0,1,0],[0,0,1]]
        broot.mainloop()

    def cylinder():
        groot.destroy()
        def calc_volume(r, h):
            return math.pi * r * r * h
        def rotate_point(x, y, z, rx, ry, rz):
            cosx, sinx = math.cos(rx), math.sin(rx)
            y, z = y * cosx - z * sinx, y * sinx + z * cosx
            cosy, siny = math.cos(ry), math.sin(ry)
            x, z = x * cosy + z * siny, -x * siny + z * cosy
            cosz, sinz = math.cos(rz), math.sin(rz)
            x, y = x * cosz - y * sinz, x * sinz + y * cosz
            return x, y, z
        def draw_cylinder(r, h, x_angle, y_angle, z_angle):
            canvas.delete("all")
            steps = 30
            rx = math.radians(x_angle)
            ry = math.radians(y_angle)
            rz = math.radians(z_angle)
            top_circle = []
            bottom_circle = []
            for i in range(steps):
                theta = (2 * math.pi / steps) * i
                x = r * math.cos(theta)
                y = r * math.sin(theta)
                top = rotate_point(x, y, h / 2, rx, ry, rz)
                bottom = rotate_point(x, y, -h / 2, rx, ry, rz)
                top_circle.append(top)
                bottom_circle.append(bottom)
            for i in range(steps):
                x1, y1, z1 = top_circle[i]
                x2, y2, z2 = bottom_circle[i]
                x1 = x1 * 5 + 200
                y1 = y1 * 5 + 200
                x2 = x2 * 5 + 200
                y2 = y2 * 5 + 200
                canvas.create_line(x1, y1, x2, y2, fill="black")
            for i in range(steps):
                j = (i + 1) % steps
                xt1, yt1, _ = top_circle[i]
                xt2, yt2, _ = top_circle[j]
                xb1, yb1, _ = bottom_circle[i]
                xb2, yb2, _ = bottom_circle[j]
                xt1, yt1 = xt1 * 5 + 200, yt1 * 5 + 200
                xt2, yt2 = xt2 * 5 + 200, yt2 * 5 + 200
                xb1, yb1 = xb1 * 5 + 200, yb1 * 5 + 200
                xb2, yb2 = xb2 * 5 + 200, yb2 * 5 + 200
                canvas.create_line(xt1, yt1, xt2, yt2, fill="blue")
                canvas.create_line(xb1, yb1, xb2, yb2, fill="blue")
        def clickvolume():
            try:
                r_val = float(r.get())
                h_val = float(h.get())
                rx, ry, rz = x_angle.get(), y_angle.get(), z_angle.get()
                volume.set(round(calc_volume(r_val, h_val), 2))
                draw_cylinder(r_val, h_val, rx, ry, rz)
            except:
                volume.set("Invalid input")
        root = Tk()
        root.geometry("1000x1000")
        root.title("Cylinder")
        root.configure(bg="lightblue")
        r = DoubleVar()
        h = DoubleVar()
        x_angle, y_angle, z_angle = IntVar(), IntVar(), IntVar()
        volume = StringVar()
        Label(root, text="Radius").pack()
        Entry(root, textvariable=r).pack()
        Label(root, text="Height").pack()
        Entry(root, textvariable=h).pack()
        Label(root, text="X Rotation Angle").pack()
        Entry(root, textvariable=x_angle).pack()
        Label(root, text="Y Rotation Angle").pack()
        Entry(root, textvariable=y_angle).pack()
        Label(root, text="Z Rotation Angle").pack()
        Entry(root, textvariable=z_angle).pack()
        Button(root, text="Calculate Volume & Draw", command=clickvolume).pack()
        Label(root, text="Volume:").pack()
        Label(root, textvariable=volume).pack()
        canvas = Canvas(root, width=750, height=750, bg="white")
        canvas.pack()
        broot.mainloop()


    def prism():
        groot.destroy()
        def calc_volume(b, h, l):
            return 0.5 * b * h * l
        def rotate(x, y, z, rx, ry, rz):
            cosx, sinx = math.cos(rx), math.sin(rx)
            cosy, siny = math.cos(ry), math.sin(ry)
            cosz, sinz = math.cos(rz), math.sin(rz)
            y, z = y * cosx - z * sinx, y * sinx + z * cosx
            x, z = x * cosy + z * siny, -x * siny + z * cosy
            x, y = x * cosz - y * sinz, x * sinz + y * cosz
            return x, y, z
        def draw_prism(b, h, l, x_angle, y_angle, z_angle):
            canvas.delete("all")
            v0 = (-b/2, -h/2, -l/2)
            v1 = (b/2, -h/2, -l/2)
            v2 = (0, h/2, -l/2)
            v3 = (-b/2, -h/2, l/2)
            v4 = (b/2, -h/2, l/2)
            v5 = (0, h/2, l/2)
            vertices = [v0, v1, v2, v3, v4, v5]
            rx = math.radians(x_angle)
            ry = math.radians(y_angle)
            rz = math.radians(z_angle)
            rotated = [rotate(x, y, z, rx, ry, rz) for x, y, z in vertices]
            projected = [[x * 5 + 200, y * 5 + 200] for x, y, z in rotated]
            edges = [
                [0, 1], [1, 2], [2, 0],
                [3, 4], [4, 5], [5, 3],
                [0, 3], [1, 4], [2, 5]
            ]
            for a, b in edges:
                canvas.create_line(projected[a][0], projected[a][1], projected[b][0], projected[b][1], fill="black")
        def clickvolume():
            try:
                b_val = float(base.get())
                h_val = float(height.get())
                l_val = float(length.get())
                rx, ry, rz = x_angle.get(), y_angle.get(), z_angle.get()
                volume.set(round(calc_volume(b_val, h_val, l_val), 2))
                draw_prism(b_val, h_val, l_val, rx, ry, rz)
            except:
                volume.set("Invalid input")
        root = Tk()
        root.geometry("1000x1000")
        root.title("Triangular Prism")
        root.configure(bg="lightblue")
        base = DoubleVar()
        height = DoubleVar()
        length = DoubleVar()
        x_angle = IntVar()
        y_angle = IntVar()
        z_angle = IntVar()
        volume = StringVar()
        Label(root, text="Base of Triangle").pack()
        Entry(root, textvariable=base).pack()
        Label(root, text="Height of Triangle").pack()
        Entry(root, textvariable=height).pack()
        Label(root, text="Length of Prism").pack()
        Entry(root, textvariable=length).pack()
        Label(root, text="X Rotation Angle").pack()
        Entry(root, textvariable=x_angle).pack()
        Label(root, text="Y Rotation Angle").pack()
        Entry(root, textvariable=y_angle).pack()
        Label(root, text="Z Rotation Angle").pack()
        Entry(root, textvariable=z_angle).pack()
        Button(root, text="Calculate Volume & Draw", command=clickvolume).pack()
        Label(root, text="Volume:").pack()
        Label(root, textvariable=volume).pack()
        canvas = Canvas(root, width=750, height=750, bg="white")
        canvas.pack()
        broot.mainloop()
        
    def pyramid():
        groot.destroy()
        def calc(s, h):
            base_area = s ** 2
            return (1/3) * base_area * h
        def apply_rotation_matrix(point, matrix):
            x, y, z = point
            new_x = matrix[0][0] * x + matrix[0][1] * y + matrix[0][2] * z
            new_y = matrix[1][0] * x + matrix[1][1] * y + matrix[1][2] * z
            new_z = matrix[2][0] * x + matrix[2][1] * y + matrix[2][2] * z
            return new_x, new_y, new_z
        def draw_pyramid(s, h, x_angle, y_angle, z_angle):
            canvas.delete("all")
            vertices = [
                [-s/2, 0, -s/2],  # base corners on y=0 plane
                [s/2, 0, -s/2],
                [s/2, 0, s/2],
                [-s/2, 0, s/2],
                [0, h, 0]         # apex at height h
            ]
            edges = [
                [0,1], [1,2], [2,3], [3,0],  # base edges
                [0,4], [1,4], [2,4], [3,4]   # side edges to apex
            ]
            rotation_matrix_x = [
                [1, 0, 0],
                [0, math.cos(math.radians(x_angle)), -math.sin(math.radians(x_angle))],
                [0, math.sin(math.radians(x_angle)), math.cos(math.radians(x_angle))]
            ]
            rotation_matrix_y = [
                [math.cos(math.radians(y_angle)), 0, math.sin(math.radians(y_angle))],
                [0, 1, 0],
                [-math.sin(math.radians(y_angle)), 0, math.cos(math.radians(y_angle))]
            ]
            rotation_matrix_z = [
                [math.cos(math.radians(z_angle)), -math.sin(math.radians(z_angle)), 0],
                [math.sin(math.radians(z_angle)), math.cos(math.radians(z_angle)), 0],
                [0, 0, 1]
            ]

            rotated_vertices = []
            for vertex in vertices:
                v = apply_rotation_matrix(vertex, rotation_matrix_x)
                v = apply_rotation_matrix(v, rotation_matrix_y)
                v = apply_rotation_matrix(v, rotation_matrix_z)
                rotated_vertices.append(v)

            projected_points = []
            for v in rotated_vertices:
                distance = 5
                x = v[0] * distance + canvas.winfo_reqwidth() / 2
                y = -v[1] * distance + canvas.winfo_reqheight() / 2  # invert y for better visual
                projected_points.append([x, y])

            for edge in edges:
                canvas.create_line(projected_points[edge[0]][0], projected_points[edge[0]][1],
                                   projected_points[edge[1]][0], projected_points[edge[1]][1], fill="black")

        def clickvolume():
            try:
                side = float(s.get())
                height = float(h.get())
                volume.set(round(calc(side, height), 3))
                x_angle_val = int(x_angle.get())
                y_angle_val = int(y_angle.get())
                z_angle_val = int(z_angle.get())
                draw_pyramid(side, height, x_angle_val, y_angle_val, z_angle_val)
            except ValueError:
                messagebox.showerror("Error", "Please enter valid numbers.")
        root = Tk()
        root.geometry("1000x1000")
        root.configure(background="Light Blue")
        root.title("3D Pyramid")
        s = StringVar()
        h = StringVar()
        x_angle = StringVar()
        y_angle = StringVar()
        z_angle = StringVar()
        Label(root, text="Enter The Side Of The Base").pack()
        Entry(root, textvariable=s).pack()
        Label(root, text="Enter The Height Of The Pyramid").pack()
        Entry(root, textvariable=h).pack()
        Label(root, text="Enter The Rotation Angle For X-Axis").pack()
        Entry(root, textvariable=x_angle).pack()
        Label(root, text="Enter The Rotation Angle For Y-Axis").pack()
        Entry(root, textvariable=y_angle).pack()
        Label(root, text="Enter The Rotation Angle For Z-Axis").pack()
        Entry(root, textvariable=z_angle).pack()
        volume = StringVar()
        Label(root, text="The Volume Of the Pyramid Is:").pack()
        Label(root, textvariable=volume).pack()
        Button(root, text="Calculate Volume & Draw Pyramid", command=clickvolume).pack()
        canvas = Canvas(root, width=750, height=750, bg="white")
        canvas.pack()
        broot.mainloop()

    def frustum():
        groot.destroy()
        def calc_volume(r1, r2, h):
            return (1/3) * math.pi * h * (r1**2 + r2**2 + r1 * r2)
        def rotate(x, y, z, rx, ry, rz):
            cosx, sinx = math.cos(rx), math.sin(rx)
            cosy, siny = math.cos(ry), math.sin(ry)
            cosz, sinz = math.cos(rz), math.sin(rz)
            y, z = y * cosx - z * sinx, y * sinx + z * cosx
            x, z = x * cosy + z * siny, -x * siny + z * cosy
            x, y = x * cosz - y * sinz, x * sinz + y * cosz
            return x, y, z
        def draw_frustum(r1, r2, h, x_angle, y_angle, z_angle):
            canvas.delete("all")
            steps = 30
            rx = math.radians(x_angle)
            ry = math.radians(y_angle)
            rz = math.radians(z_angle)
            top = []
            bottom = []
            for i in range(steps):
                theta = (2 * math.pi / steps) * i
                x_top = r2 * math.cos(theta)
                y_top = r2 * math.sin(theta)
                x_bot = r1 * math.cos(theta)
                y_bot = r1 * math.sin(theta)
                top.append(rotate(x_top, y_top, h/2, rx, ry, rz))
                bottom.append(rotate(x_bot, y_bot, -h/2, rx, ry, rz))
            for i in range(steps):
                x1, y1, _ = top[i]
                x2, y2, _ = bottom[i]
                x1, y1 = x1 * 5 + 200, y1 * 5 + 200
                x2, y2 = x2 * 5 + 200, y2 * 5 + 200
                canvas.create_line(x1, y1, x2, y2, fill="black")
            for i in range(steps):
                j = (i + 1) % steps
                xt1, yt1, _ = top[i]
                xt2, yt2, _ = top[j]
                xb1, yb1, _ = bottom[i]
                xb2, yb2, _ = bottom[j]
                canvas.create_line(xt1*5+200, yt1*5+200, xt2*5+200, yt2*5+200, fill="blue")
                canvas.create_line(xb1*5+200, yb1*5+200, xb2*5+200, yb2*5+200, fill="blue")
        def clickvolume():
            try:
                r1_val = float(r1.get())
                r2_val = float(r2.get())
                h_val = float(h.get())
                rx, ry, rz = x_angle.get(), y_angle.get(), z_angle.get()
                volume.set(round(calc_volume(r1_val, r2_val, h_val), 2))
                draw_frustum(r1_val, r2_val, h_val, rx, ry, rz)
            except:
                volume.set("Invalid input")
        root = Tk()
        root.geometry("1000x1000")
        root.title("Frustum")
        root.configure(bg="lightblue")
        r1 = DoubleVar()
        r2 = DoubleVar()
        h = DoubleVar()
        x_angle = IntVar()
        y_angle = IntVar()
        z_angle = IntVar()
        volume = StringVar()
        Label(root, text="Bottom Radius").pack()
        Entry(root, textvariable=r1).pack()
        Label(root, text="Top Radius").pack()
        Entry(root, textvariable=r2).pack()
        Label(root, text="Height").pack()
        Entry(root, textvariable=h).pack()
        Label(root, text="X Rotation Angle").pack()
        Entry(root, textvariable=x_angle).pack()
        Label(root, text="Y Rotation Angle").pack()
        Entry(root, textvariable=y_angle).pack()
        Label(root, text="Z Rotation Angle").pack()
        Entry(root, textvariable=z_angle).pack()
        Button(root, text="Calculate Volume & Draw", command=clickvolume).pack()
        Label(root, text="Volume:").pack()
        Label(root, textvariable=volume).pack()
        canvas = Canvas(root, width=750, height=750, bg="white")
        canvas.pack()
        broot.mainloop()

    def tetrahedron():
        groot.destroy()
        def calc_volume(a):
            return (a ** 3) / (6 * math.sqrt(2))
        def rotate(x, y, z, rx, ry, rz):
            cosx, sinx = math.cos(rx), math.sin(rx)
            cosy, siny = math.cos(ry), math.sin(ry)
            cosz, sinz = math.cos(rz), math.sin(rz)
            y, z = y * cosx - z * sinx, y * sinx + z * cosx
            x, z = x * cosy + z * siny, -x * siny + z * cosy
            x, y = x * cosz - y * sinz, x * sinz + y * cosz
            return x, y, z
        def draw_tetrahedron(a, x_angle, y_angle, z_angle):
            canvas.delete("all")
            h = math.sqrt(2 / 3) * a
            vertices = [
                (0, 0, h/2),  # Top
                (-a/2, -math.sqrt(3)*a/6, -h/2),
                (a/2, -math.sqrt(3)*a/6, -h/2),
                (0, math.sqrt(3)*a/3, -h/2)
            ]
            rx = math.radians(x_angle)
            ry = math.radians(y_angle)
            rz = math.radians(z_angle)
            rotated = [rotate(x, y, z, rx, ry, rz) for x, y, z in vertices]
            projected = [[x * 5 + 200, y * 5 + 200] for x, y, z in rotated]
            edges = [
                [0, 1], [0, 2], [0, 3],
                [1, 2], [2, 3], [3, 1]
            ]
            for a, b in edges:
                canvas.create_line(projected[a][0], projected[a][1], projected[b][0], projected[b][1], fill="black")
        def clickvolume():
            try:
                a_val = float(side.get())
                rx, ry, rz = x_angle.get(), y_angle.get(), z_angle.get()
                volume.set(round(calc_volume(a_val), 2))
                draw_tetrahedron(a_val, rx, ry, rz)
            except:
                volume.set("Invalid input")
        root = Tk()
        root.geometry("1000x1000")
        root.title("Tetrahedron")
        root.configure(bg="lightblue")
        side = DoubleVar()
        x_angle = IntVar()
        y_angle = IntVar()
        z_angle = IntVar()
        volume = StringVar()
        Label(root, text="Side Length").pack()
        Entry(root, textvariable=side).pack()
        Label(root, text="X Rotation Angle").pack()
        Entry(root, textvariable=x_angle).pack()
        Label(root, text="Y Rotation Angle").pack()
        Entry(root, textvariable=y_angle).pack()
        Label(root, text="Z Rotation Angle").pack()
        Entry(root, textvariable=z_angle).pack()
        Button(root, text="Calculate Volume & Draw", command=clickvolume).pack()
        Label(root, text="Volume:").pack()
        Label(root, textvariable=volume).pack()
        canvas = Canvas(root, width=750, height=750, bg="white")
        canvas.pack()
        broot.mainloop()

    cbutton=Button(groot,text="Cube",command=cube,width=10,height=2).place(x=50,y=50)
    cdbutton=Button(groot,text="Cuboid",command=cuboid,width=10,height=2).place(x=300,y=50)
    sbutton=Button(groot,text="Sphere",command=sphere,width=10,height=2).place(x=50,y=150)
    cyhbutton=Button(groot,text="Cylinder",command=cylinder,width=10,height=2).place(x=300,y=150)
    pbutton=Button(groot,text="Prism",command=prism,width=10,height=2).place(x=50,y=250)
    pybutton=Button(groot,text="Pyramid",command=pyramid,width=10,height=2).place(x=300,y=250)
    frbutton=Button(groot,text="Frustum",command=frustum,width=10,height=2).place(x=50,y=350)
    tebutton=Button(groot,text="Tetrahedron",command=tetrahedron,width=10,height=2).place(x=300,y=350)

button1=Button(root,text="2-D Figures",command=twod,width=15,height=3,bg="black",fg="white").place(x=100,y=225)
button2=Button(root,text="3-D Figures",command=thrd,width=15,height=3,bg="black",fg="white").place(x=400,y=225)

root.mainloop()
```
